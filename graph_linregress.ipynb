```python
# Importing dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
qld_covid = pd.read_csv("qld_covid_data.csv")
vic_covid = pd.read_csv("vic_covid_data.csv")
wa_covid = pd.read_csv("wa_covid_data.csv")
```

```python
qld_covid
```

```python
qld_df = pd.read_csv("dee/Untitled Folder/qld_percent_change_australia.csv")
qld_df = qld_df.iloc[10:]
qld_df
```

```python
wa_df = pd.read_csv("dee/Untitled Folder/wa_percent_change_australia.csv")
wa_df = wa_df.iloc[10:]
wa_df
```

```python
vic_df = pd.read_csv("dee/Untitled Folder/vic_percent_change_australia.csv")
vic_df = vic_df.iloc[10:]
vic_df
```

```python
plt.figure(figsize=((20,20)))
x_val = qld_covid["percentage"]
y_val = qld_df["diff"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val, y_val)
regress_values = x_val * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
r = "The r-value is: " + str(rvalue)
plt.annotate(r,(-100,-2.5),fontsize=15,color="red")
plt.annotate(line_eq,(-100,-3),fontsize=15,color="red")
plt.xlabel("Percentage Change of Covid-19 in QLD")
plt.ylabel("Percentage Change of STP Index in QLD")
plt.plot(x_val,regress_values,"r-")
plt.scatter(qld_covid["percentage"],qld_df["diff"])
print(f"The r-squared value is: {rvalue**2}")
plt.savefig("qld_corr.png")
```

```python
plt.figure(figsize=((15,15)))
x_val = vic_covid["percentage"]
y_val = vic_df["diff"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val, y_val)
regress_values = x_val * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
r = "The r-value is: " + str(rvalue)
plt.annotate(r,(-100,-2.75),fontsize=15,color="red")
plt.annotate(line_eq,(-100,-3),fontsize=15,color="red")
plt.xlabel("Percentage Change of Covid-19 in VIC")
plt.ylabel("Percentage Change of STP Index in VIC")
plt.plot(x_val,regress_values,"r-")
plt.scatter(vic_covid["percentage"],vic_df["diff"])
print(f"The r-squared value is: {rvalue**2}")
plt.savefig("vic_corr.png")
```

```python
# Importing dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
# Read in csv files
qld_covid = pd.read_csv("qld_covid_data.csv")
vic_covid_newcases = pd.read_csv("vic_new_covid_data.csv")
vic_covid = pd.read_csv("vic_covid_data.csv")
wa_covid = pd.read_csv("wa_covid_data.csv")
food_df = pd.read_csv("food_accommodation_stp.csv")
```

```python
qld_df = pd.read_csv("qld_percent_change_australia.csv")
# qld_df = qld_df.iloc[10:]
```

```python
wa_df = pd.read_csv("wa_percent_change_australia.csv")
# wa_df = wa_df.iloc[10:]
```

```python
vic_df = pd.read_csv("vic_percent_change_australia.csv")
vic_df = vic_df.iloc[10:]
```

# Scatterplots and linear regressions

## Queensland - Active cases

```python
# Generate a scatter plot and perform a linear regression of percent changes in STP index and active covid cases
x_val = qld_covid["percentage"]
y_val = qld_df["diff"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val, y_val)
regress_values = x_val * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
r = "The r-value is: " + str(rvalue)
rsq = "The r-squared value is: " + str(rvalue**2)

plt.annotate(rsq,(-200,-2),fontsize=10,color="red")
plt.annotate(r,(-200,-1.5),fontsize=10,color="red")
plt.annotate(line_eq,(-200,-1),fontsize=10,color="red")

plt.xlabel("Percentage Change of Covid-19 in QLD")
plt.ylabel("Percentage Change of STP Index in QLD")
plt.plot(x_val,regress_values,"r-")
plt.scatter(qld_covid["percentage"],qld_df["diff"])
print(f"The r-squared value is: {rvalue**2}")
plt.tight_layout()
plt.savefig("qld_corr_final.png")
```
